```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


np.random.seed(42)
dates = pd.date_range('2025-01-01', periods=100)
price = 100 + np.cumsum(np.random.randn(100))
df = pd.DataFrame({'date': dates, 'price': price})

df['MA7'] = df['price'].rolling(window=7).mean()
df['MA21'] = df['price'].rolling(window=21).mean()

plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['price'], label='Cena akcji', color='blue')
plt.plot(df['date'], df['MA7'], label='MA7', color='orange')
plt.plot(df['date'], df['MA21'], label='MA21', color='green')
plt.fill_between(df['date'], df['MA7'], df['MA21'], where=(df['MA7'] > df['MA21']), color='lightcoral', alpha=0.5, label='MA7 > MA21')
plt.fill_between(df['date'], df['MA7'], df['MA21'], where=(df['MA7'] <= df['MA21']), color='lightblue', alpha=0.5, label='MA7 <= MA21')
plt.title('Wykres ceny akcji z MA7 i MA21')
plt.xlabel('Data')
plt.ylabel('Cena')
plt.legend()
plt.grid()
plt.show()  
```
