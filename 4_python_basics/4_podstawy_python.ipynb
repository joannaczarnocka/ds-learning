Zadanie 9 – Normalizacja Min-Max

```python
min_val = 50
max_val = 100

def normalize(value, min_val, max_val):
    return (value - min_val) / (max_val - min_val)

def check_included_in_0_1(normalized):
    return 0 <= normalized <= 1

for value in [75, 50, 100, 62.5]:
    normalized = normalize(value, min_val, max_val)
    included_in_0_1 = check_included_in_0_1(normalized)
    print(f"Value: {value}, Normalized: {normalized}, Included in [0, 1]: {included_in_0_1}")

print("\nTable:")  

for value in [75, 50, 100, 62.5]:
    normalized = normalize(value, min_val, max_val)
    print (f"{value} | {normalized}")
```

Zadanie 11 – Sprawdzanie poprawności hiperparametrów

```python
learning_rate = 0.01
batch_size = 64
epochs = 100
dropout = 0.5

# Sprawdź czy learning_rate w zakresie (0, 1]
if 0 < learning_rate <= 1:
    print("OK")
else:
    print(f"ERROR: Learning rate {learning_rate} is NOT in (0,1]")
    
# Sprawdź czy batch_size jest potęgą 2 (użyj operacji bitowych lub logarytmów)
import math

def is_power_of_two_log(n):
    return n > 0 and math.log2(n).is_integer()

if is_power_of_two_log(batch_size):
    print("OK")
else:
    print(f"ERROR: Batch size {batch_size} is NOT power of two")


# Sprawdź czy epochs > 0 i < 1000
if 0 < epochs <= 1000:
    print("OK")
else:       
    print(f"ERROR: Epochs {epochs} is NOT in (0,1000)") 

# Sprawdź czy dropout w zakresie [0, 1]
if 0 <= dropout <= 1:
    print("OK")
else:
    print(f"ERROR: Dropout {dropout} is NOT in [0,1]")
```

Zadanie 12 – Estymacja czasu treningu

```python
number_of_samples = 50000

batch_size = 128

epochs = 50

time_per_batch_seconds = 0.15

full_batches = number_of_samples // batch_size

remainder = number_of_samples % batch_size

batches_per_epoch = full_batches + (1 if remainder > 0 else 0)

total_time_seconds = batches_per_epoch * epochs * time_per_batch_seconds
print(f"Time: {total_time_seconds} s")

hours = total_time_seconds // 3600
remaining = total_time_seconds % 3600
minutes = remaining // 60
seconds = remaining % 60

print(f"Szacowany czas: {hours} h {minutes} min {seconds} s")
```
