Zadanie 12 - numpy

```python
point_a = np.array([1, 2, 3, 4, 5])
point_b = np.array([5, 4, 3, 2, 1])
points = np.random.rand(4, 5)
def euclidean_distance(a, b):
    return round(np.sqrt(np.sum((a - b) ** 2)),2)    
distance_ab = euclidean_distance(point_a, point_b)
distance_ab
```

```python
distance_matrix = np.zeros((4, 4))
for i in range(4):
    for j in range(4):
        distance_matrix[i, j] = euclidean_distance(points[i], points[j])
distance_matrix
```

Zadanie 9 - pandas

```python
chunk_size = 100
age_means = []
for i, chunk in enumerate(pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv", chunksize=chunk_size)):
    mean_age = chunk['Age'].mean()
    age_means.append(mean_age)
    print(f"Batch {i+1}: Mean Age = {mean_age}")    
global_mean = round(sum(age_means) / len(age_means),2)
print(f"Global Mean Age = {global_mean}")

global_mean_no_chunks = round(titanic['Age'].mean(), 2)
print(f"Global Mean Age without chunks = {global_mean_no_chunks}")
```
