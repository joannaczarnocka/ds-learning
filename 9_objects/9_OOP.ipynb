```python
class  BankAccount:
    def __init__(self, wlasciciel , saldo):
        self.wlasciciel = wlasciciel
        self.saldo = saldo

    def wplac(self, kwota):
        self.saldo += kwota

    def wyplac(self, kwota):
        self.saldo -= kwota

    def pokaz_saldo(self):
        print(self.saldo)

ba = BankAccount('Jan Kowalski', 20)
ba.wplac(10)
ba.wyplac(1)
ba.pokaz_saldo() 
```

```python
class Student:  
    def __init__(self, imie, numer_indeksu):
        self.imie = imie
        self.numer_indeksu = numer_indeksu
        self.oceny = []

    def dodaj_ocene(self, ocena):
        self.oceny.append(ocena)

    def srednia(self):
        if not self.oceny:
            return 0
        return sum(self.oceny) / len(self.oceny)

    def __str__(self):
        return f'Student: {self.imie}, Numer indeksu: {self.numer_indeksu}, Oceny: {self.oceny}, Średnia: {self.srednia():.2f}'


student1 = Student("Edmund Kowalski", "12345") 
student2 = Student("John Smith", "98765")
student3 = Student("Alice Cat", "100000")

student1.dodaj_ocene(5)
student1.dodaj_ocene(4)
student1.dodaj_ocene(3)

student2.dodaj_ocene(2)
student2.dodaj_ocene(3)
student2.dodaj_ocene(4)

student3.dodaj_ocene(5)
student3.dodaj_ocene(5)
student3.dodaj_ocene(4)
student3.dodaj_ocene(4)

print(student1)
print(student2)
print(student3)

    
```

```python
class Uczen:
    def __init__(self, imie, oceny=[]):
        self.imie = imie
        self.oceny = []
        self.oceny.extend(oceny)

class Klasa:
    def __init__(self, nazwa, uczniowie=[]):
        self.nazwa = nazwa
        self.uczniowie = []
        self.uczniowie.extend(uczniowie)

    def dodaj_ucznia(self, uczen):
        self.uczniowie.append(uczen)

    def srednia_klasy(self):
        if not self.uczniowie:
            return 0
        suma_ocen = 0
        liczba_ocen = 0     
        for uczen in self.uczniowie:
            suma_ocen += sum(uczen.oceny)
            liczba_ocen += len(uczen.oceny)
        if liczba_ocen == 0:
            return 0
        return suma_ocen / liczba_ocen

    def najlepszy_uczen(self):
        najlepszy = None
        najwyzsza_srednia = -1
        for uczen in self.uczniowie:
            if len(uczen.oceny) == 0:
                continue
            srednia_ucznia = sum(uczen.oceny) / len(uczen.oceny)
            if srednia_ucznia > najwyzsza_srednia:
                najwyzsza_srednia = srednia_ucznia
                najlepszy = uczen
        return najlepszy
    
uczen1 = Uczen("John", [5, 4, 3])   
uczen2 = Uczen("Mary", [2, 3, 4])
uczen3 = Uczen("David", [5, 5, 4, 4])
uczen4 = Uczen("Eric", [3, 4, 2])
uczen5 = Uczen("Robert", [4, 4, 5])

klasa = Klasa("4B")

klasa.dodaj_ucznia(uczen1)
klasa.dodaj_ucznia(uczen2) 
klasa.dodaj_ucznia(uczen3)
klasa.dodaj_ucznia(uczen4)
klasa.dodaj_ucznia(uczen5) 

print(f'Średnia klasy {klasa.nazwa}: {klasa.srednia_klasy():.2f}')
najlepszy = klasa.najlepszy_uczen()
if najlepszy:
    print(f'Najlepszy uczeń: {najlepszy.imie} ze średnią {(sum(najlepszy.oceny) / len(najlepszy.oceny)):.2f}')
```

```python
class Akcja:
    def __init__(self, nazwa, cena_kupna, cena_aktualna):
        self.nazwa = nazwa
        self.cena_kupna = cena_kupna
        self.cena_aktualna = cena_aktualna
        self.ilosc = 0
        
class Portfel:
    def __init__(self, kapital_poczatkowy, akcje = []):
        self.kapital_poczatkowy = kapital_poczatkowy
        self.akcje = []
        self.akcje.extend(akcje)

    def kup_akcje(self, akcja, ilosc):
        akcja.ilosc += ilosc
        self.akcje.append(akcja)

    def sprzedaj_akcje(self, akcja, ilosc):
        if akcja.ilosc >= ilosc:
            akcja.ilosc -= ilosc

    def aktualizuj_ceny(self, slownik_cen):
        for akcja in self.akcje:
            if akcja.nazwa in slownik_cen:
                akcja.cena_aktualna = slownik_cen[akcja.nazwa]

    def wartosc_portfela(self):
        wartosc = 0
        for akcja in self.akcje:
            wartosc += akcja.cena_aktualna * akcja.ilosc
        return wartosc

    def zysk_strata(self):
        wartosc_obecna = self.wartosc_portfela()
        return wartosc_obecna - self.kapital_poczatkowy


portfel = Portfel(10000)
akcja1 = Akcja("A1", 150, 155)
akcja2 = Akcja("B2", 200, 250)
akcja3 = Akcja("C3", 300, 340)
akcja4 = Akcja("D4", 400, 480)
akcja5 = Akcja("E5", 500, 560)            

portfel.kup_akcje(akcja1, 10)
portfel.kup_akcje(akcja2, 2)        
portfel.kup_akcje(akcja3, 1)
portfel.kup_akcje(akcja4, 5)
portfel.kup_akcje(akcja5, 3)
portfel.sprzedaj_akcje(akcja4, 2)

nowe_ceny = {
    "A1": 1156, 
    "B2": 1251,
    "C3": 1341,
    "D4": 1481,
    "E5": 1561
}
portfel.aktualizuj_ceny(nowe_ceny)  
print(f'Wartość portfela: {portfel.wartosc_portfela():.2f}')
print(f'Zysk/Strata: {portfel.zysk_strata():.2f}')          
```
