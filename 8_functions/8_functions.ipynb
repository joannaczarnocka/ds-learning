Zadanie 18 – Funkcja kroswalidacji 

```python
def k_fold_split(dane, k):
    fold_size = len(dane) // k
    return [dane[i*fold_size:(i+1)*fold_size] for i in range(k)] 

dane = list(range(1, 101))
k = 5
folds = k_fold_split(dane, k)
for i, fold in enumerate(folds):
    print(f"Fold {i+1}: {fold}")    
```
