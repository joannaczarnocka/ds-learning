```python
import numpy as np
import matplotlib.pyplot as plt

learning_rates = [0.001, 0.01, 0.1, 0.5 , 1.0, 1.5]
iterations = 30 
start_point = 5

def function(x):
    """Funkcja do zminimalizowania: f(x) = x²"""
    return x ** 2

def gradient(x):
    """Gradient (pochodna) funkcji: f'(x) = 2x"""
    return 2 * x

def gradient_descent(start_x, learning_rate, num_iterations):
    """
    Algorytm spadku gradientu.
    Parametry:
    - start_x: punkt startowy
    - learning_rate: współczynnik uczenia (η)
    - num_iterations: liczba iteracji
    Zwraca:
    - historia wartości x i f(x)
    """
    x = start_x
    history = {'x': [x], 'f': [function(x)]}
    for i in range(num_iterations):
    # Obliczenie gradientu
        grad = gradient(x)
        # Aktualizacja x: x = x - η * gradient
        x = x - learning_rate * grad
        # Zapisanie historii
        history['x'].append(x)
        history['f'].append(function(x))
        if i < 5 or i == num_iterations - 1:
            print(f"Iteracja {i+1}: x = {x:.6f}, f(x) = {function(x):.6f}, gradient = {grad:.6f}")

    return history

# Uruchomienie gradient descent
print("=== Gradient Descent dla f(x) = x² ===\n")
for lr in learning_rates:
    print(f"\n--- Learning Rate: {lr} ---")

    history = gradient_descent(
        start_x=start_point, 
        learning_rate=lr, # Współczynnik uczenia
        num_iterations=iterations # Liczba kroków
        )
    # Wizualizacja
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    x_range = np.linspace(-6, 6, 100)
    axes[0].plot(x_range, function(x_range), 'b-', linewidth=2, label='f(x) = x²')
    axes[0].scatter(history['x'], history['f'], c=range(len(history['x'])),
    cmap='Reds', s=100, zorder=5, edgecolors='black')
    axes[0].plot(history['x'], history['f'], 'r--', alpha=0.5)
    axes[0].scatter([history['x'][0]], [history['f'][0]], c='green', s=200,
    marker='*', zorder=6, label='Start')
    axes[0].scatter([history['x'][-1]], [history['f'][-1]], c='red', s=200,
    marker='*', zorder=6, label='Koniec')
    axes[0].set_xlabel('x')
    axes[0].set_ylabel('f(x)')
    axes[0].set_title('Gradient Descent: trajektoria')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    # Wykres 2: Wartość funkcji w czasie
    axes[1].plot(history['f'], 'b-o', linewidth=2, markersize=6)
    axes[1].set_xlabel('Iteracja')
    axes[1].set_ylabel('f(x)')
    axes[1].set_title('Spadek wartości funkcji straty')
    axes[1].grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    print(f"\nOptymalne x: {history['x'][-1]:.6f} (powinno być 0)")
    print(f"Minimalna wartość f(x): {history['f'][-1]:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression


np.random.seed(42)
X, y = make_regression(n_samples=200, n_features=1, noise=20, random_state=42)
X = X.flatten()
print(f"Dane: {len(X)} obserwacji")
print(f"X: zakres [{X.min():.2f}, {X.max():.2f}]")
print(f"y: zakres [{y.min():.2f}, {y.max():.2f}]")

def predict(X, w, b):
    """Predykcja: y = wX + b"""
    return w * X + b

def compute_mse(y_true, y_pred):
    """Obliczenie MSE"""
    return np.mean((y_true - y_pred) ** 2)

def compute_gradients(X, y, w, b):
    """
    Obliczenie gradientów dla regresji liniowej.
    MSE = (1/n) * Σ(y - (wx + b))²
    ∂MSE/∂w = (-2/n) * Σ(y - (wx + b)) * x
    ∂MSE/∂b = (-2/n) * Σ(y - (wx + b))
    """
    n = len(y)
    y_pred = predict(X, w, b)
    error = y - y_pred
    dw = (-2/n) * np.sum(error * X)
    db = (-2/n) * np.sum(error)
    return dw, db

def gradient_descent_linear(X, y, learning_rate=0.01, num_iterations=1000):
    """
    Gradient Descent dla regresji liniowej.
    """
    # Inicjalizacja wag (losowa)
    w = np.random.randn()
    b = np.random.randn()
    history = {
    'w': [w], 'b': [b],
    'loss': [compute_mse(y, predict(X, w, b))]
    }
    for i in range(num_iterations):
        # Obliczenie gradientów
        dw, db = compute_gradients(X, y, w, b)
        # Aktualizacja parametrów
        w = w - learning_rate * dw
        b = b - learning_rate * db
        # Zapisanie historii
        loss = compute_mse(y, predict(X, w, b))
        history['w'].append(w)
        history['b'].append(b)
        history['loss'].append(loss)
        if (i + 1) % 200 == 0:
            print(f"Iteracja {i+1}: w = {w:.4f}, b = {b:.4f}, MSE = {loss:.4f}")
    return w, b, history

# Trening
print("\n=== Trening Gradient Descent ===\n")
w_final, b_final, history = gradient_descent_linear(X, y, learning_rate=0.1, num_iterations=1000)
print(f"\nWyuczone parametry: w = {w_final:.4f}, b = {b_final:.4f}")
print(f"Końcowy MSE: {history['loss'][-1]:.4f}")
# Wizualizacja
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
# 1. Dane i linia regresji
axes[0].scatter(X, y, alpha=0.6, label='Dane')
x_line = np.linspace(X.min(), X.max(), 100)
axes[0].plot(x_line, predict(x_line, w_final, b_final), 'r-',
linewidth=2, label=f'y = {w_final:.2f}x + {b_final:.2f}')
axes[0].set_xlabel('X')
axes[0].set_ylabel('y')
axes[0].set_title('Regresja liniowa (Gradient Descent)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
# 2. Krzywa uczenia (Loss)
axes[1].plot(history['loss'], 'b-', linewidth=2)
# Schemat Gradient Descent
axes[1].set_xlabel('Iteracja')
axes[1].set_ylabel('MSE (Loss)')
axes[1].set_title('Krzywa uczenia')
axes[1].set_yscale('log')
axes[1].grid(True, alpha=0.3)
# 3. Ewolucja parametrów
axes[2].plot(history['w'], label='w (waga)', linewidth=2)
axes[2].plot(history['b'], label='b (bias)', linewidth=2)
axes[2].set_xlabel('Iteracja')
axes[2].set_ylabel('Wartość parametru')
axes[2].set_title('Ewolucja parametrów')
axes[2].legend()
axes[2].grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
# Porównanie ze sklearn
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X.reshape(-1, 1), y)
print(f"\nSklearn LinearRegression: w = {lr.coef_[0]:.4f}, b ={lr.intercept_:.4f}")




```
