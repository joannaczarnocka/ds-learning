```python
studenci = [
    ('Anna', 'Kowalska', 5.0, 4.5, 5.0),
    ('Jan', 'Nowak', 3.0, 3.5, 3.0),
    ('Maria', 'Zielińska', 4.0, 4.5, 5.0),
    ('Piotr', 'Wiśniewski', 2.0, 3.0, 2.5),
    ('Ewa', 'Wójcik', 5.0, 5.0, 4.5),
    ('Tomasz', 'Kamiński', 4.0, 3.5, 4.0)
]                                                           


kategorie = {
    'wyróżniający': [],
    'dobry': [],
    'średni': []
}

for student in studenci:
    imie, nazwisko, *oceny = student
    srednia = sum(oceny) / len(oceny)
    student_info = (imie, nazwisko, round(srednia, 2))
    
    if srednia >= 4.5:
        kategorie['wyróżniający'].append(student_info)
    elif srednia >= 3.5:
        kategorie['dobry'].append(student_info)
    else:
        kategorie['średni'].append(student_info)

for kategoria, lista_studentow in kategorie.items():
    print(f"{kategoria.capitalize()}:")
    for student in lista_studentow:
        print(f"  {student}")

print("\nPogrupowani studenci:", kategorie)
```

```python
dzien1 = {'user001', 'user002', 'user003', 'user004', 'user005', 'user010'}
dzien2 = {'user002', 'user003', 'user006', 'user007', 'user008', 'user010'}
dzien3 = {'user003', 'user004', 'user009', 'user010', 'user011'}    

wszystkie_3_dni = dzien1 & dzien2 & dzien3
tylko_dzien1 = dzien1 - (dzien2 | dzien3)
tylko_dzien2 = dzien2 - (dzien1 | dzien3)
tylko_dzien3 = dzien3 - (dzien1 | dzien2)
dokladnie_2_dni = ((dzien1 & dzien2) | (dzien1 & dzien3) | (dzien2 & dzien3)) - wszystkie_3_dni
aktywnosc_uzytkownikow = {
    'wszystkie_3_dni': wszystkie_3_dni,
    'tylko_dzien1': tylko_dzien1,
    'tylko_dzien2': tylko_dzien2,
    'tylko_dzien3': tylko_dzien3,
    'dokladnie_2_dni': dokladnie_2_dni
}   
print("\nUżytkownicy aktywni wszystkie 3 dni:", wszystkie_3_dni)
print("Użytkownicy tylko w dniu 1:", tylko_dzien1)
print("Użytkownicy tylko w dniu 2:", tylko_dzien2)
print("Użytkownicy tylko w dniu 3:", tylko_dzien3)
print("Użytkownicy aktywni dokładnie 2 dni:", dokladnie_2_dni)

raport_aktywnosci = {}
for uzytkownik in dzien1 | dzien2 | dzien3:
    dni_aktywnosci = sum([
        uzytkownik in dzien1,
        uzytkownik in dzien2,
        uzytkownik in dzien3
    ])
    raport_aktywnosci[uzytkownik] = dni_aktywnosci  
print("\nRaport aktywności użytkowników:")
for uzytkownik, dni in raport_aktywnosci.items():
    print(f"  {uzytkownik}: {dni} dni")

```

```python
filmy_gatunki = {
    'Matrix': ['Akcja', 'Sci-Fi'],
    'Shrek': ['Komedia', 'Familijny', 'Animacja'],
    'Inception': ['Akcja', 'Sci-Fi', 'Thriller'],
    'Toy Story': ['Animacja', 'Familijny'],
    'Die Hard': ['Akcja', 'Thriller'],
    'The Hangover': ['Komedia']
}

gatunki_filmy = {}
for film, gatunki in filmy_gatunki.items():
    for gatunek in gatunki:
        gatunki_filmy.setdefault(gatunek, set()).add(film)

print("\nGatunki i odpowiadające im filmy:")
for gatunek, filmy in gatunki_filmy.items():
    print(f"  {gatunek}: {filmy}")
```
